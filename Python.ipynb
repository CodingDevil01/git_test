**Fct d'appartenance d'un element a une liste**

```python
def appartenance(elt,lst):
    for i in lst:
        if i==elt:
            return True
    return False
print(appartenance(2, [5, 4, 1, 2, 3])) #True
print(appartenance(6, [5, 4, 1, 2, 3])) #False
```

**Fct qui renvoit l'indice d'elt si il appartient a la lst**

```python
def indice(elt,lst):
    for i in lst:
        if i==elt:
            return lst.index(elt)
    return None
print(indice(2, [5, 4, 1, 2, 3]))
print(indice(6, [5, 4, 1, 2, 3]))
```

## **Recherche d’un élément dans une liste triée(Dichotomie)**

Lorsque l’on dispose d’une liste triée par ordre croissant, on peut grandement améliorer notre algorithme en utilisant le principe de **dichotomie**.

On recherche tout d’abord l’élément central de la liste.

Si c’est l’élément recherché, on s’arrête. Sinon, on le compare à l’élément recherché.

Si l’élément recherché est inférieur à l’élément central, on le recherche dans la première partie de la liste. Sinon, on le recherche dans la deuxième partie de la liste.

On retourne donc à la première étape de notre algorithme appliqué à l’une des deux demi-listes.

```python
def appartient_dicho(elt,lst):
    g=0
    d=len(lst)-1
    while g<=d:
        m=(g+d)//2
        if lst[m]==elt:
            return True
        if elt<lst[m]:
            d=m-1
        else:
            g=m+1
    return False
print(appartient_dicho(13, [1, 3, 5, 7, 8, 10, 13, 14, 17, 19]))
print(appartient_dicho(18, [1, 3, 5, 7, 8, 10, 13, 14, 17, 19]))
```
